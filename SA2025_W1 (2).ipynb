### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
df = pd.read_csv('Cars.csv')

median_mpg = df['mpg'].median()

filtered_cars = df[df['mpg'] > median_mpg].copy()

filtered_cars['hp_per_weight'] = filtered_cars['horsepower'] / filtered_cars['weight']

max_ratio = filtered_cars['hp_per_weight'].max()

best_cars = filtered_cars[filtered_cars['hp_per_weight'] == max_ratio]

print("Car(s) with the best horsepower-to-weight ratio among cars with above-median mpg:")
print(best_cars[['name', 'horsepower', 'weight', 'mpg', 'hp_per_weight']])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

pivot_df = avg_mpg.pivot(index='model_year', columns='origin', values='mpg')

plt.figure(figsize=(10, 6))
for origin in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[origin], marker='o', label=origin.capitalize())

plt.title('Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Y = weight

Color by: origin

Size by: mpg

Hue order = ['japan', 'europe', 'usa']

Add meaningful plot titles and axis titles.

```python
#your code here
plt.figure(figsize=(12, 8))
scatter = sns.scatterplot(
    data=df,
    x='horsepower',
    y='weight',
    hue='origin',
    hue_order=['japan', 'europe', 'usa'],
    size='mpg',
    sizes=(20, 200),
    alpha=0.7,
    palette='deep'
)

plt.title('Car Weight vs Horsepower Colored by Origin and Sized by MPG', fontsize=16)
plt.xlabel('Horsepower', fontsize=14)
plt.ylabel('Weight (lbs)', fontsize=14)

plt.legend(title='Origin & MPG Size', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
car_counts = df['name'].value_counts()
multi_year_cars = car_counts[car_counts > 1].index.tolist()

multi_year_df = df[df['name'].isin(multi_year_cars)]

mpg_stats = multi_year_df.groupby('name')['mpg'].agg(['std', 'mean', 'count'])

consistent_cars = mpg_stats[(mpg_stats['std'] < 1.0) & (mpg_stats['count'] >= 2)]

consistent_cars_sorted = consistent_cars.sort_values(by=['count', 'mean'], ascending=[False, False])

print("Consistent cars (std(mpg) < 1.0 and multiple years):")
print(consistent_cars_sorted.rename(columns={'std': 'std_mpg', 'mean': 'avg_mpg', 'count': 'appearances'}))
```
